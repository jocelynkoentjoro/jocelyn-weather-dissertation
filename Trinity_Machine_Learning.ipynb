# Machine Learning Prediction

  REFERENCE : https://github.com/Hrishikesh332/Weather-Forecasting-Using-ML/tree/main
  
  Adjustment made to the desired dataset and problem:
  1. Importing data not from kaggle, but from my own data through connect google drive
  2. Skip data pre-processing because we don't use any labeled weather data
  3. Adding more performance metrics: MSE, RMSE, R2 SCORE

```python
from sklearn.model_selection import train_test_split

features=["temperature_2m_max (°C)", "temperature_2m_min (°C)", "wind_gusts_10m_max (m/s)", "wind_speed_10m_max (m/s)","precipitation_sum (mm)","relative_humidity_2m_mean (%)","surface_pressure_mean (hPa)"]
X=df[features]
y=df["temperature_2m_mean (°C)"]
train_X, test_X, train_y, test_y = train_test_split(X, y,random_state = 0)
```

# Decision Tree Regressor:

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import root_mean_squared_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

model_decision_tree=DecisionTreeRegressor(random_state=1)
model_decision_tree.fit(train_X, train_y)
pred_decision_tree=model_decision_tree.predict(test_X)
print("MAE (Mean Absolute Error): %f" %(mean_absolute_error(test_y, pred_decision_tree)))
print("RMSE (Root Mean Squared Error): %f" %(root_mean_squared_error(test_y, pred_decision_tree)))
print("MSE (Mean Squared Error): %f" %(mean_squared_error(test_y, pred_decision_tree)))
print("R2 Score: %f" %(r2_score(test_y, pred_decision_tree)))
```

# Random Forest Regressor:

```python
from sklearn.ensemble import RandomForestRegressor

model_random_forest=RandomForestRegressor(random_state=1)
model_random_forest.fit(train_X, train_y)
pred_random_forest=model_random_forest.predict(test_X)
print("MAE (Mean Absolute Error): %f" %(mean_absolute_error(test_y, pred_random_forest)))
print("RMSE (Root Mean Squared Error): %f" %(root_mean_squared_error(test_y, pred_random_forest)))
print("MSE (Mean Squared Error): %f" %(mean_squared_error(test_y, pred_random_forest)))
print("R2 Score: %f" %(r2_score(test_y, pred_random_forest)))
```

# Extreme Gradient Regressor:

```python
from xgboost import XGBRegressor

model_xgb= XGBRegressor(n_estimators=100, learning_rate=0.04)
model_xgb.fit(train_X, train_y)
pred_xgb=model_xgb.predict(test_X)
print("MAE (Mean Absolute Error): %f" %(mean_absolute_error(test_y, pred_xgb)))
print("RMSE (Root Mean Squared Error): %f" %(root_mean_squared_error(test_y, pred_xgb)))
print("MSE (Mean Squared Error): %f" %(mean_squared_error(test_y, pred_xgb)))
print("R2 Score: %f" %(r2_score(test_y, pred_xgb)))
```

# Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingRegressor

model_gb= GradientBoostingRegressor(random_state=1)
model_gb.fit(train_X, train_y)
pred_gb=model_gb.predict(test_X)
print("MAE (Mean Absolute Error): %f" %(mean_absolute_error(test_y, pred_gb)))
print("RMSE (Root Mean Squared Error): %f" %(root_mean_squared_error(test_y, pred_gb)))
print("MSE (Mean Squared Error): %f" %(mean_squared_error(test_y, pred_gb)))
print("R2 Score: %f" %(r2_score(test_y, pred_gb)))
```

# 5x2cv Paired T-test
REFERENCE : https://rasbt.github.io/mlxtend/user_guide/evaluate/paired_ttest_5x2cv/

# Linear Regression vs Decision Tree

```python
from mlxtend.evaluate import paired_ttest_5x2cv

linear_regression_score = model_linear_regression.fit(train_X, train_y).score(test_X, test_y)
decision_tree_score = model_decision_tree.fit(train_X, train_y).score(test_X, test_y)

#check model score
print('Linear Regression Accuracy: %.2f%%' % (linear_regression_score*100))
print('Decision Tree Accuracy: %.2f%%' % (decision_tree_score*100))

#assume that
a = 0.05
t, p = paired_ttest_5x2cv(estimator1=model_linear_regression,
                          estimator2=model_decision_tree,
                          X=X, y=y,
                          random_seed=1)

#print t statistics and p value
print('t statistic: %.3f' % t)
print('p value: %.3f' % p)

#summary
if ( p < a):
  print ("There is a significant difference between two models")
else :
  print ("Both models perform well with the dataset")
```
